## A/B-тестирование

А/В-тест - это эксперимент с двумя группами для определения лучшего из двух вариантов. Если используется стандартный вариант (или же он отсутствует), то он называется _контрольным_. Типичная гипотеза состоит в том, что предлагаемый вариант лучше контрольного.

__Вариант эксперемента (treatment)__ - нечто, что предлагается испытуемому в целях тестирования.

__Тестовая группа (treatment group)__ - группа испытуемых, которым которым предлагается конкретный вариант.

__Контрольная группа (control group)__ - группа испытуемых, которым предлагается стандартный вариант или не предлагается никакой.

__Рандомизация (randomization)__ - процесс случайного отнесения испытуемых к вариантам.

__Испытуемые (subjects)__ - субъекты (пациенты, посетители веб-сайта), которым предлагаются варианты.

__Проверочная статистика (test statistics)__ - метрика, которая используется для измерения эффекта варианта.

Проверочная статистика, или метрика, которая используется для сравнения группы А с группой В, наиболее встречающейся метрикой является двоичная переменная.

Без контрольной группы нет никакой гарантии, что "прочие условия будут равными", а любая разница действительно обусловлена вариантом эксперемента (либо случайным образом). Контрольная группа, подчиняется тем же условиям (за исключением интересующего варианта), что и тестовая группа.

_Слепое исследование_ - это исследование, в котором испытуемые не осведомлены о том, что им предлагается вариант А либо вариант В. В двойном слепом исследовании исследователи и помощники не осведомлены о том, какие испытуемые учавствуют и какой вариант им предлагается.

Необходимо серьезно обеспечивать сохранность принципов рандомизации. Одна метрика или проверочная статистика, должна быть установлена заранее. Выбор проверочной статистики после того, как эксперемент проведен, приводит к смещению вследствии предвзятости исследователя.

На вопрос:"Какая из много численных возможных цен является лучшей?" используется относительно новый этап планирования эксперемента: _многорукий бандит_.

## Проверка значимости

_Проверки гипотез (проверка значимости)_ в статистическом анализе предназначены для того, чтобы узнать, может ли случайность быть отвественной за наблюдаемый эффект. Статистическая проверка гипотез была изобретена как способ защитить исследователей от того, чтобы оказаться обманутым случайностью.

__Неправильное истолкование случайности__

Склонность людей недооценивать случайность можно наблюдать в следующем эксперименте. Попросите нескольких своих друзей вообразить 50-кратное подбрасывание монеты: пусть они запишут серию случайных О (орел) и Р (решка). Затем попросите их, чтобы они на самом деле подбросили монету 50 раз и записали результаты. Пусть они поместят реальные результаты подбрасывания монеты в один список и выдуманные результаты — в другой. Легко различить, какие результаты являются настоящими: настоящие результаты будут иметь более длинную вереницу, состоящую из О или Р. Увидеть в наборе из 50 настоящих подбрасываний пять или шесть О или Р 
подряд не является чем-то необычным. Однако, когда большинство из нас воображает случайные подбрасывания монеты, и у нас получается три или четыре О подряд, мы говорим себе: чтобы серия выглядела случайной, нам лучше всего переключиться на Р. Другая сторона этой монеты, если можно так выразиться, состоит в том, что когда мы на деле видим реальный эквивалент, состоящий из шести О подряд (например, когда один заголовок объявления превосходит по результативности другой на 10%), мы склонны приписать это чему-то реальному, не просто случайности.

__Нулевая гипотеза (null hypothesis)__ - гипотеза о том, что виной всему является случайность. Потребуется доказательство того, что разница между группами является более предельной, чем та которая обоснованно могла бы быть порождена случайностью. Любая разница между группами обусловлена случайностью - это базисное допущение и называется нулевой гипотезой. Природа нулевой гипотезы определяет структуру проверки гипотезы. Цель исследований показать неправильность нулевой гипотезы и показать, что исходы для групп А и В различаются больше, чем то, что может породить случайность. Объедененные перетасованные результаты из групп А и В и процедура повторного отбора из них воплощают нулевую гипотезу о том, что группы А и В являются эквивалентными и взаимозаменяемыми, и называется _нулевой моделью_. 

__Альтернативная гипотеза (alternative hypothesis)__ - гипотеза, компенсирующая нулевую (то, что надеемся доказать). Нулевая и альтернативная гипотезы составляют полное множество всех событий.

__Односторонняя проверка (one-way test)__ - проверка значимости, при которой случайных результатов подсчитывается только в одном направлении.

__Двусторонняя проверка (two-way test)__ - проверка гипотезы, при которой количество случайных результатов подсчитывается в двух направлениях.

Нередко в А/В-тесте проверяется новая возможность (например В) относительно взятой по умолчанию возможности (А) и изначально исходим из того, что придерживаемся исходной возможности, если только новая возможность не окажется определенно лучше. В таком случае при проверке гипотез потребуется защититься от того, чтобы не быть обманутым случайностью по направлению в пользу В. Не волнует то, что можно быть обманутым случайностью в другом направлении. Поэтому хочется иметь альтернативную гипотезу (В лучше А). В таком случае используется проверка односторонней  гипотезы (или гипотезу с одним хвостом). Это означает, что предельный шанс приводит только в однонаправленному подсчету в сторону р-значения.

Если требуется, чтобы проверка гипотезы защитила исследователя от того, чтобы быть обманутым случайностью в любом направлении, то альтернативная гипотеза должна быть _двунаправленной_ (А отличается от В и может быть больше или меньше). В таком случае используется _двустороняя гипотеза_ (или гипотеза с двумя хвостами). Это означает предельный шанс приводит к двунаправленному подсчету в сторону р-значения.

Тема различий между проверками с одним хвостом или с двумя хвостами является довольно запутанной и не имеет прямого отношения к науке о данных, где прецизионность расчетов р-значения не особо важна.

## Повторный отбор

_Повторный отбор_ в статистике означает многократное извлечение значений из наблюдаемых данных с общей целью определения случайной вариабельности  в статистической величине. Также используется для определения и улучшения точности некоторых автоматически обучающихся моделей (например, модель на основе дерева решений, построенная на многократно бутстрапированных наборах данных, которые усредняется в резельтате процесса, именуемого __бэггингом__).

Два главных типа процедур повторного отбора: _бутстрап_ и _перестановка_ (пермутация). Перестановочные тесты применяется для проверки гипотез и в типичной ситуации предусматривают две или более групп.

__Перестановочный тест (permutation test)__ - (рандомизационный тест, случайный пермутационный тест, точный тест) процедура объединения двух более выборок и случайное (либо исчерпывающее) перераспределение наблюдений в повторные выборки. В процедуре _перестановки_ задействуются две или более выборки. Перестановка означает изменение порядка следования значений. Первый шаг в _перестановочной проверке_ гипотезы состоит в объединении результатов из групп А и В (групп С и D...). В этом заключается логическое воплощение нулевой гипотезы - варианты эксперемента, которые были предложены группами, не различаются. Проверяется гипотеза путем случайного извлечения групп из этого объединенного множества и определяется насколько отличаются друг от друга группы. Процедура перестановке такова:
1. Объединить результаты из разных групп в единный набор данных;
2. Перетасовать объединенные данные, затем в случайном порядке извлечь (без возврата) повторную выборку того же размера, что и группа А (очевидно, что она будет содержать немного данных из других групп);
3. Из оставшихся данных в случайном порядке извлечь (без возврата)  повторную выборку того же размера, что и группа В;
4. Сделать тоже самое для групп C, D и т.д. Собран один набор  повторных выборок, которые отржают размеры исходных выборок;
5. В зависимости от статистике или оценки, которая была вычислена для исходных выборок (например, разница в групповых долях), теперь рассчитать ее для повторных выборок и записать: это будет одной итерацией перестановки;
6. Повторить предыдущие шаги R раз для получения перестановочного распределения проверочной статистике.

Сравним наблюдаемую разницу между группами с набором перестановленных разниц. Вопрос состоит в том, находится ли эта разница внутри диапазона того, что может быть порождено случайностью, или же, напротив, является статистически значимой? Если наблюдаемая разница убедительно лежит в пределах набора  перестановленных разниц, то ничего не доказано - наблюдаемая разница находится внутри диапазона того, что может породить случайность. Если наблюдаемая разница лежит вне большей части перестановочного распределения, то приходим к заключению, что случайность не несет ответственности и разница является _статистически значимой_.

_Цель_ перестановочного теста - объединить все группы тестов на проверку разницы между группами, порождена ли эта разность случайностью.

__Повторный отбор (resampling)__ - извлечение дополнительных выборок ("повторных выборок") из наблюдаемого набора данных.

__С возвратом либо без возврата (with or without replcament)__ - при отборе элемент либо возвращается, либо не возвращается в выборку перед следующим извлечением.

__Эрзац-переменная (прокси)__ - (косвенная переменная) это переменная, которая подменяет истинную интересующую переменную, возмножно отсутствующую, либо слишком дорогостоящую, либо измерение которой требует слишком продолжительного времени. При этом полезно иметь по крайней мере немного данных об истинной переменной, являющейся предметом интереся , чтобы определить силу ее связи с эрзацем.

```python
session_times = pd.read_csv('data/web_page_data.csv')
session_times.Time = 100 * session_times.Time
```

```python
# Функция boxplot испольщует именнованый арумент by, чтобы создать рисунок
ax = session_times.boxplot(by='Page', column='Time',
                           figsize=(4, 4))
ax.set_xlabel('')
ax.set_ylabel('Time (in seconds)')
plt.suptitle('')

plt.tight_layout()
plt.show()
```

```python
mean_a = session_times[session_times.Page == 'Page A'].Time.mean()
mean_b = session_times[session_times.Page == 'Page B'].Time.mean()
print(mean_b - mean_a)
```

```python
# функция, которая случайно назначает 36 времен сеансов группе из 21 элемента и группе из 15 элементов
def perm_fun(x, nA, nB):
    """Вызов этой функции в количестве R=1000 и установка nA=21 и nB=15 приводят к распределению разниц во временах сеансов"""
    n = nA + nB
    idx_B = set(random.sample(range(n), nB))
    idx_A = set(range(n)) - idx_B
    return x.loc[list(idx_B)].mean() - x.loc[list(idx_A)].mean()
    return x.loc[idx_B].mean() - x.loc[idx_A].mean()
```

```python
nA = session_times[session_times.Page == 'Page A'].shape[0]
nB = session_times[session_times.Page == 'Page B'].shape[0]
print(perm_fun(session_times.Time, nA, nB))
```

```python
random.seed(1)
perm_diffs = [perm_fun(session_times.Time, nA, nB) for _ in range(1000)]

fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_diffs, bins=11, rwidth=0.9)
ax.axvline(x = mean_b - mean_a, color='black', lw=2)
ax.text(50, 190, 'Observed\ndifference', bbox={'facecolor':'white'})
ax.set_xlabel('Session time differences (in seconds)')
ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# convert perm_diffs to numpy array to avoid problems with some Python installations
perm_diffs = np.array(perm_diffs)
print(np.mean(perm_diffs > mean_b - mean_a))
```

Полученный процент 12,1% свидетельствует о том, что наблюдаемая разница (во временах сеансов между страницей А и В) находится далеко внутри диапазона случайной вариации и, следовательно, не является статистически значимой.

__Исчерпывающий и бутстраповский перестановочные тесты__

В дополнение к предыдущей процедуре _случайной перестановки_, также именнуемой __случайным перестановочным тестом__ или __рандомизационным тестом__, существуют два варианта перестановочного теста:
* исчерпывающий перестановочный тест - вместо простой случайной перестановки и разделения данных, фактически выясняем все возможные способы, которыми они могут быть разделены. Это осуществимо для малых размеров выборок. При большом числе повторных перестановок результаты случаного перестановочного теста апроксимируют результаты исчерпывающего перестановочного теста и приближаются к ним в пределе. Исчерпывающие перестановочные тесты также иногда именуются точными тестами из-за их статистического свойства, которое гарантирует, что нулевая модель не протестируется как "значимая" больше альфа-уровня теста.
* бутстраповский перестановочный тесты - извлечения делаются с возвратом, благодаря этомупроцедура повторного отбора моделирует не только случайный элемент в названчении варианта испытуемому, но и случайный элемент в отборе испытуемых из популяции. 

Обе процедуры встречаются в статистике, и разница между ними носит довольно замысловатый характер и не имеет особой важности в практике о науке о данных.

Перестановочные тесты являются полезными эвритическими процедурами для разведывания роли случайной вариации. Они также легко предлагают нетривиальный окольный путь в обход формализма и "ложного детерминизма" формализованных статистик.

Одно из достоинств повторного отбора состоит в том, что он почти вплотную сближается с шаблонным подходом к статистическому выводу.

## Статистическая значимость и р-значения

_Статистическая значимость_ описывает дает ли эксперемент более предельный результат, чем тот, который может быть получен случайностью. Если результат лежит за пределами случайной вариации, то принято говорить, что он является _статистически значимым_.

__р-Значение (p-value)__ - с учетом случайной модели, которая воплощает нулевую гипотезу, р-значение является вероятностью получения результатов столь же необычных или предельных, что и наблюдаемые результаты. р-Значение - это частота, с которой случайная модель производит результат, более предельный, чем наблюдаемый. р-Значение представляет на самом деле: вероятность, что с учетом случайной модели, могут произойти настолько же предельные результаты, что и наблюдаемые.

Шесть принципов для исследователей от Американской статистической ассоциации (American Statistical Association, ASA):
1. р-Значения могут указывать на то, насколько несовместны данные с заданной статистической моделью;
2. р-Значения не измеряют вероятность, что изучаемая гипотеза является истиной, либо вероятность, что данные были произведены исключительно в силу случайности;
3. Научные выводы и деловые или стратегические решения не должны базироваться только на том, переходит ли р-значение определенный порог;
4. Налдежащий статистический вывод требует полной отчетности и прозрачности;
5. р-Значение, или статистическая значимость, не измеряет размер эффекта или важности результата;
6. Как таковое р-значение нне обеспечивает хорошую меру фактических данных относительно модели или гипотезы.

Если результат является статистически значимым, это не означает, что он имеет практическую значимость. Крупные выборки усиливают малые, несущественные эффекты настолько, чтобы исключить случайность в объяснении. Когда р-значение недотягивает до статистической значимости (она превышает 5%), это интерпритируется как "эффект не доказан". Может оказаться, что более крупная выборка произведет меньшее р-значение.

Основная функция проверок значимости (проверка гипотез) - защитить от того, чтобы быть обманутым случайностью. Они структурируются таким образом, чтобы минимизировать ошибки 1-го рода.

Для исследователя данных р-значение полезная метрика в ситуациях, где стоит знать, лежит ли модельный результат, который кажется интересным и полезным, в диапазоне нормальной случайной выриабельности.

Проверк значимости используются для определения того, лежит ли наблюдаемый эффект внутри диапазона случайно вариации для модели нулевой гипотезы. р-Значение - это вероятность, которая приводит к результатам, таким же предельным, какми могут оказаться наблюдаемые результаты с учетом модели нулевой гипотезы. Значение альфы - это порог "необычности" в случайной модели нулевой гипотезы. 

__Альфа (alpha)__ - (уровень значимости) вероятностный порог "необычности", который должен превзойти случайные результаты, чтобы фактические исходы считались статистически значимыми. Перед эксперементом устанавливается порог, обычно " более предельный, чем 5% случайных результатов (нулевая гипотеза)": этот порог называется _альфой_. "Какова вероятность настолько предельного результата с  учетом случайной модели" - на такой вопрос отвечает р-значение и альфа устанавливается порогом.

__Ошибка 1-го рода (type 1 error)__ - ошибочный вывод о том, что эффект является реальным тогда, как он обусдовден случайностью.

__Ошибка 2-го рода (type 2 error)__ - ошибочный вывод о том, что эффект обусловлен случайностью, тогда как он является реальным. Ошибка 2-го рода является больше суждением, что размер выборки слишком мал, чтобы можно было обнаружить эффект. 

```python
random.seed(1)
obs_pct_diff = 100 * (200 / 23739 - 182 / 22588)
print(f'Observed difference: {obs_pct_diff:.4f}%')
conversion = [0] * 45945
conversion.extend([1] * 382)
conversion = pd.Series(conversion)

perm_diffs = [100 * perm_fun(conversion, 23739, 22588) 
              for _ in range(1000)]

fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_diffs, bins=11, rwidth=0.9)
ax.axvline(x=obs_pct_diff, color='black', lw=2)
ax.text(0.06, 200, 'Observed\ndifference', bbox={'facecolor':'white'})
ax.set_xlabel('Conversion rate (percent)')
ax.set_ylabel('Frequency')
ax.set_title('Частотное распределение для разницы в интенсивностях конверсии между страницами А и В')

plt.tight_layout()
plt.show()
```

```python
print(np.mean([diff > obs_pct_diff for diff in perm_diffs]))

survivors = np.array([[200, 23739 - 200], [182, 22588 - 182]])
chi2, p_value, df, _ = stats.chi2_contingency(survivors)

print(f'p-значение для одностороннего теста: {p_value / 2:.4f}')
```

##  Проверки на основе t-статистики

Существуют многочисленные типы проверок значимости, которые зависят от того, являются ли данные количественными или измеряемыми, сколько имеется выборок и что измеряется. Самой общепринятой ялвется _t-тест_ на основе _t-статистики_, отсылающей к _t-распределению Стьюдента_, первоначально разработанной У.С.Госсетом для аппроксимации распределения одновыборочного среднего.

__Проверочная статистика (test statistics)__ - метрика для интересующей разницы или эффекта.

__t-Статистика (t-statistics)__ - стандартизированная версия часто встречающихся проверочных статистик, таких как среднее значение.

__t-Распределение (t-distribution)__ - эталонное распределение (в данном случае полученное из нулевой гипотезы), с которым может быть сопоставлена наблюдаемая t-статистика.

В тесте на основе повторного отбора шкала данных не играет роли. Хорошей аппроксимацией перестановочного распределения является t-тест. Для того, чтобы t-распределение можно было применять без привязки к какой-либо конкретной шкале данных, должна использоваться стандартизированная форма проверочной статистики. 

```python
res = stats.ttest_ind(session_times[session_times.Page == 'Page A'].Time, 
                      session_times[session_times.Page == 'Page B'].Time,
                      equal_var=False)
print(f'p-значение для одностороннего теста: {res.pvalue / 2:.4f}')
```

```python
tstat, pvalue, df = sm.stats.ttest_ind(
    session_times[session_times.Page == 'Page A'].Time, 
    session_times[session_times.Page == 'Page B'].Time,
    usevar='unequal', alternative='smaller')
print(f'p-value: {pvalue:.4f}')
```

## Множественное тестирование

Множественное тетирование - https://www.stat.berkeley.edu/~mgoldman/Section0402.pdf

__Ошибка 1-го рода (type 1 error)__ - Ошибочный вывод, что эффект является статистически значимым.

__Коэффициент ложных открытий (false discovery rate)__ - доля совершения ошибки 1-го рода в рузультате множественного тестирования. Данный термин изначальн использовался для описания интенсивности, с которой данный набор проверок гипотез ложно выявляет значимый эффект.

__Инфляция альфы (alpha inflation)__ - Явление множественного тестирования, в котором альфа, вероятность ошибки 1-го рода, увеличивается по мере проведения большого числа тестов.

__Корректировка p-значений (adjustment of p-value)__ - уточнение значения при выполнении множественного тестирования на одинаковых данных.

__Переподгонка (overfitting)__ - подготовка к шуму.

__Инфляция альфы__ - предсказательные и переменная исхода сгенерированны случайным образом, проведя 20 проверок значимости с альфа уровнем 0.05, одна из переменных окажется ошибочно стат. значимой (ошибка 1-го рода). Эту вероятность можно рассчитать так: вероятность, что одна из переменных пройдет проверку равна 0.95; вероятность того, что все 20 предсказателей пройдут проверку равна $0.95^{20} = 0.36$. Вероятность того, что, что по ркайней мере один предсказатель ложно покажет значимость равна 0.64.

Чем больше переменных добавляется или больше болей выполняется, тем выше вероятность того, что нечто проявится как "значимое" по чистой случайности.

Корректировочные процедуры в статистике могут компенсировать факт быть обманутым случайностью путем установки более строгой планки для статистической значимости, чем планка для одной проверки гипотезы. Такие корректировочные процедуры, как правило, предусматривают "деление альфы" по числу проверок. Данный метод приводит к меньше альфа-уровню (т.е. более строгой планке для статистической значимости) для каждой проверки. Корректировка Бонферрони делит альфу на число сравнений. HSD (honest significant difference, честной значимой разницей) Тьюки применяют к максимальной разнице между групповыми средними, срванивая ее с эталоном, основанным на t-распределении.

Тест Даннета - https://davidmlane.com/hyperstat/B112114.html

## Степени свободы

Понятие "степени свободы" применяется к статистикам, вычисляемым из выборочных данных и относятся к числу значений, которые могут вырьироваться свободно. (степень свободы часто озгначает n-1)

__n или размер выборки (n or sample size)__ - число наблюдений (строк или записей) в данных.

__d.f. (degrees of freedom)__ - степень свободы.

Традиционные курсы или учебники статистики учат разнообразным стандартным проверкам гипотез (проверки на основе t-статистики, F-статистики и т.д.).

## Дисперсионный анализ

Статистическая процедура, которая выполняет проверку на статистическуи значимую разницу среди групп, называется _дисперсионным анализом_ (Analysis Of Variance, ANOVA).

__Попарное сравнение (pairwise comparison)__ -проверка гипотезы (к примеру, о средних значениях) между двумя группами из множества групп.

__Универсальный тест (omnibus test)__ - одна-единственная проверка гипотезы о совокупной дисперсии среди средних значений иноэества групп.

__Разложение дисперсии (decomposition of variance)__ - выделение компонентов, которые вносят вклад в отдельное значение (к примеру, из совокупного среднего, из среднего значения варианта эксперемента и из остаточной ошибки).

__F-статистика (F-statistic)__ - стандартизированнная статистика. измеряющая степень, с которой разницы в групповых средних превышают то, что можно ожидать в случайной модели.

__Сумма квадратов (SS, sum of squares)__ - обозначает отклонение от какого-либо среднего значения.

Чем больше делать попарных сравнений переменных в выборке, тем больше потенциал для того, чтобы оказаться обманутым случайностью. Можно выполнить всего один совокупный тест, который дает ответ на вопрос: могут ли все переменные иметь в своей основе одинаковую гипотезу, и разница между переменными обусловлена случайным образом. 

Для проверки такого вопроса используется процедура дисперсионного анализа. Основополагающие принципы ее применения можно увидеть в процедуре повторного отбора:
1. Объединить все данные;
2. Перетасовать и извлечь повторные выборки (количество как было) с тем же количеством значений;
3. Записать среднее значение каждой из групп;
4. Записать дисперсию среди средних значений этих групп;
5. Повторить шаги 2-4 много раз (например, 1000).

```python
four_sessions = pd.read_csv("data/four_sessions.csv")

ax = four_sessions.boxplot(by='Page', column='Time',
                           figsize=(4, 4))
ax.set_xlabel('Page')
ax.set_ylabel('Time (in seconds)')
plt.suptitle('')
plt.title('')

plt.tight_layout()
plt.show()
```

```python
print(pd.read_csv("data/four_sessions.csv").head())
```

```python
observed_variance = four_sessions.groupby('Page').mean().var()[0]
print('Observed means:', four_sessions.groupby('Page').mean().values.ravel())
print('Variance:', observed_variance)
# Permutation test example with stickiness
def perm_test(df):
    df = df.copy()
    df['Time'] = np.random.permutation(df['Time'].values)
    return df.groupby('Page').mean().var()[0]
    
print(perm_test(four_sessions))

random.seed(1)
perm_variance = [perm_test(four_sessions) for _ in range(3000)]
print('Pr(Prob)', np.mean([var > observed_variance for var in perm_variance]))

fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(perm_variance, bins=11, rwidth=0.9)
ax.axvline(x = observed_variance, color='black', lw=2)
ax.text(60, 200, 'Observed\nvariance', bbox={'facecolor':'white'})
ax.set_xlabel('Variance')
ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## F-статистика

Для дисперсионного анализа существует статистическая проверка на основе **F-статистики**, которая опирается на отношение дисперсии по всем групповым средним (т.е. вариантного эффекта) к дисперсии вследствие остаточной ошибки. Чем выше это отношение, тем более статистически значимым является результат.

Пакет **statsmodels** предоставляет имплементацию процедуры дисперсионного анализа на Python.

```python
model = smf.ols('Time ~ Page', data=four_sessions).fit()
                
aov_table = sm.stats.anova_lm(model)
print(aov_table)
```

, где DF-это степень свободы, Sum Sq-сумма квадратов, Mean Sq-средние квадраты (среднеквадратические отклонения), F value - F-статистика.

Для общего среднего сумма квадратов - это отступление общего среднего от 0, возведенное в квадрат и умноженное на 20 (по числу наблюдений). Параметр степеней свободы для общего среднего по определению равен 1.

Для остатков степени свободы равны 20 (все наблюдения могут варьироваться), и SS - это сумма квадратичных разниц между отдельными наблюдениями и вариантными средними. Средние квадраты (MS) - это сумма квадратов, деленная на степени свободы.

F-статистика равна MS(вариант)/MS(ошибка).

__Разложение дисперсии__

Любое наблюдаемое внутри набора данных значение можно разбить на общее среднее, вариантный эффект и остаточную ошибку. Данная процедура называется __разложением дисперсии__.
1. Начать с общего среднего;
2. Добавить вариантный эффект, который может быть отрицательными;
3. Добавить остаточную ошибку, которая может быть отрицательной.

## Двухсторонний дисперсионный анализ

Разложение дисперсии для левого верхнего значения в таблице A-B-C-D-теста таково:
* Начать с общего среднего (173,5);
* Добавить вариантный (групповой) эффект (-1,75 [172-173,75]);
* Добавить остаток (-8 [164-172]);
* Получаем (164).

Только что описанный тест является односторонним дисперсионным анализом, в котором имеется один варьирующийся фактор (группу). Работать с двухсторонним дисперсионным анализом, как с односторонним дисперсионным анализом путем выявления "эффекта взаимодействия".

Регулярный дисперсионный анализ и затем двухсторонний дисперсионный анализ являются первыми шагами по дороге к полной статистической модели, такой как регрессия и логистическая регрессия.

__Ключевые идеи__ для дисперсионного анализа:
* Дисперсионный анализ (ANOVA) - это статистическая процедура для анализа результатов эксперемента с многочисленными группами.
* Указанная процедура является расширением аналогичных процедур для A/B-теста, используемая для определения, находится ли совокупная вариация среди групп внутри диапазона случайной вариации.
* Полезным результатом дисперсионного анализа служит выявление диспрсионных компонентов, ассоциированных с групповыми вариантами, эффектами взаимодействия и ошибками.
