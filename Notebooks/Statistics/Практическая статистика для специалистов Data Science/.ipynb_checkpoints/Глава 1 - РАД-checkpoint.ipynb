## Элементы структурированных данных

Одна из наиболее часто встречающихся форм структурированных данных является _таблица_ со строками и столбцами. Два базовых типа структурированных данных:
1. Числовые данные (numeric) - данные, которые выражаются на числовой шкале:
    * дискретные (discrete, счетная величина, целое число) - данные, которые могут принимать только целочисленные значения, такие как количества;
    * непрерывные (интервал, числовая величина, число с плавающей точкой) - данные, которые могут принимать любое значение в интервале;
2. Категориальные данные (categorical, фиксированный набор значений) - данные, которые могут принимать только конкретное множество значений, представляющих множество возможных категорий:
    * двоичные (binary) - частный случай категориальных данных с двумя категориями значений;
    * порядковые (ordinal, ранговые) - данные, которые имеют явное заданное упорядочение.

## Прямоугольные данные

__Прямоугольные данные__ - это общий термин для двумерной матрицы, в которой строки обозначают _записи (records, случаи, примеры, прецеденты, экземпляр, наблюдения, шаблоны, патерны, образец)_, а столбцы - _признаки (features, переменные, атрибуты, предикаторы, предсказатели)_. __Кадр данных (data frame)__ - это базовая структура данных для статистических и автоматически обучающихся моделей. Это специфичный формат, присущий языкам R и Python. __Исход (outcome)__ - зависимая переменная, выход, отклик, результат.

В статистике и науке о данных используются разные термины. Для статистиков предсказательные переменные, которые применяются для предсказания _отклика_ либо зависимой переменной. Для исследователя данных признаки, используются для предсказания цели. _sample_ - обозначается одна-единственная строка, для статистика _sample_ - коллекция строк, т.е. выборка.

Традиционные таблицы базы данных имеют столбец-индекс. 

__Непрямоугольные структуры данных__ - это данные _временного ряда_, _пространственные структуры данных_, _графовые (или сетевые) структуры данных_.

## Оценки центрального положения

__Центральное положение__ - одна из размерностей в обобщении признака.

__Среднее (mean)__ $\bar{x}$ - (среднее арифметическое) сумма всех значений, деленная на число значений. $$\bar{x} = \frac{\sum_{i=1}^n x_i}{n}$$

__Среднее взвешенное (weighted mean)__ $\bar{x}_w$ - (срденее арифметическое взвешенное) сумма произведений всех зачений на их веса, деленная на сумму весов. $$\bar{x}_w = \frac{\sum_{i=1}^n w_i x_i}{\sum_{i=1}^n w_i}$$

__Среднее усеченное (trimmed mean)__ $\bar{x}$ - (обрезанное среднее) среднее число всех значений после отбрасывания фиксированного числа дальних значений. Вычисляется путем отбрасывания фиксированного числа сортированных значений $x_{(1)}, x_{(2)},..., x_{(n)}$, где $x_{(1)}$ - наименьшее значение, $x_{(n)}$ - наибольшее значение. $$\bar{x} = \frac{\sum_{i=p+1}^{n-p}x_{(i)}}{n-2p}$$ Усеченное среднее устраняет влияние предельных значений. Среднее усеченное может считаться компромиссом между медианой и средним: оно робастно к предельным значениям в данных, но использует больше данных для расчета оценки центрального положения.

__Медиана (median)__ - (50-й процентиль) такое значение, половина сортированных данных находится выше и ниже данного значения, т.е. такое число, расположенное в сортированном списке данных ровно посередине.

__Медиана взвешенная (weighted mediana)__ - такое значение, что половина суммы весов находится выше и ниже сортированных данных, т.е. такое значение, в котором сумма весов равна для нижней и верхней половин сортированного списка. 

Медиана и взвешенная медиана робастны к выбросам.

__Робастный (robust)__ - (устойчивый) не чувствительный к предельным значениям.

__Выброс (outlier)__ - (предельное, экстремальное, анамальное значение) значение данных, которое сильно отличается от большинства данных, т.е. это любое значение, которое сильно дистанцировано от других значений в наборе данных.

В статистике была разработанна масса других робастных метрик центрального положения, так называемых эстиматоров.


В статистике применяется термин _оценка_, для исследователей данных это _метрические показатели_ или _метрики_.

```python
import pandas as pd
import numpy as np
from scipy.stats import trim_mean
from statsmodels import robust

import seaborn as sns
import matplotlib.pylab as plt
```

```python
state = pd.read_csv("data/state.csv")
```

```python
# среднее значение или среднеe арифметическое
state.Population.mean()
```

```python
# среднее усеченное значение
trim_mean(state['Population'], 0.1) # 0.1 - отбросить по 10% с каждого конца
```

```python
# медиана
state['Population'].median()
```

```python
# взвешенное среднее
np.average(state['Murder.Rate'], weights=state['Population'])
```

```python
import wquantiles

# взвешенная медиана
wquantiles.median(state['Murder.Rate'], weights=state['Population'])
```

## Оценка вариабельности

__Оценка вариабельности__ или __дисперсность__ - вторая размерность, показывающая плотность сгруппированности данных или их разброс. Широко используемые оценки вариабельности основаны на разницах, или отклонениях, между оценкой центрального положения и наблюдаемыми данными.

__Отклонения (deviations)__ - (ошибки, погрешности, остатки) разница между наблюдаемыми значениями и оценкой центрального положения.

__Дисперсия (variance)__ $s^2, \sigma^2$ - (среднеквадратическое отклонение, среднеквадратическая ошибка) сумма квадратичных отклонений от среднего, деленная на степени свободы (n-1), где n - число значений данных. $$s^2 = \frac{\sum(x-\bar{x})^2}{n-1}$$ Оценка является смещенной если делить на _n_, при _n-1_ оценка будет несмещенной.

__Стандартное отклонение (standard deviation)__ $\sigma, s$ - (норма $l_2$, евклидова норма) квадратный корень из дисперсии. $$s = \sqrt{\sigma^2}$$

__Среднее абсолютное отклонение (mean absolute deviation)__ - (норма $l_1$, манхэттенская норма) среднее абсолютных значений отклонений от среднего. $$mad = \frac{\sum_{i=1}^n |x_i - \bar{x}|}{n}$$, где $\bar{x}$ - среднее значение в выборке, или выборочное среднее.

Дисперсия, стандартное отклонение, среднее абсолютное отклонение не устойчивы к выбросам и предельным значениям.

__Медианное абсолютное отклонение от медианы (median absolute deviation from the median)__ MAD - медиана абсолютных значений отклонений от медианы. MAD является робастной оценкой. $$MAD = median(|x_1 - m|, |x_2 - m|,...,|x_N - m|)$$, m - медиана $x_1, x_2,...,x_n$. Иногда MAD умножают на постоянный поправочный коэффициент, чтобы поставить метрику  в ту же шкалу, что и стандартное отклонение. Часто используемый фактор 1,4826 означает. что 50% нормального распределения попадает в диапазон $\pm$MAD.

Дисперсия, стандартное оклонение, среднее абсолютное отклонение и медианное абсолютное отклонение от медианы не являются эквивалентными. Стандартное отклонение всегда больше среднего абсолютного отклонения. котрое в свою очередь, больше медианного абсолютного отклонения.

__Размах (range)__ - разница между максимальным и минимальным значением в наборе данных. Размах черезвучайно чувствителен к выбросам.

Статистические показатели на основе сортированных (ранжированных) данных  называются _порядковыми статистиками_. __Порядковые статистики (order statistics)__ - (ранги) метрики на основе значений данных, отсортированных по возрастанию.

__Процентиль (precentile)__ - (квантиль) такое значение, что Р процент значений принимает данное значение или меньшее и (100-Р) процент значений принимает данное значение или большее. Процентиль по существу, аналогичен _квантилю_.

__Межквартальный размах (interquartile range)__ - (МКР, IQR) разница между 75-м м 25-м процентилями.

```python
# стандартное отклонение или норма  l1
state['Population'].std()
```

```python
# IQR = 3-й квантиль - 1-й квантиль
state['Population'].quantile(0.75) - state['Population'].quantile(0.25)
```

```python
# MAD медианное абсолютное отклонение от медианы
robust.scale.mad(state['Population'])
```

```python
robust.scale.mad(state['Population']) * 1.4826
```

В статистической теории _центральное положение_ и _вариабельность_ упоминаются как _моменты распределения первого и второго порядка_. Моменты третьего и четвертого порядка - это _ассиметрия_ и _эксцесс_. __Ассиметрия__ - смещение данных к большим или меньшим значениям, __эксцесс__ - склонность данных к предельным значениям.

## Разведывание распределения данных

__Коробчатая диаграмма (boxplot)__ - (диаграмма типа "ящик с усами") график, введенный в употребление Тьюки в качестве быстрого способа визуализации распределения данных. Коробчатые диаграммы основаны на процентилях.

```python
ax = (state['Population']/1_000_000).plot.box()
ax.set_ylabel('Население, млн человек')
```

На графике видно, что медианная численность населения штатов составляют 5млн, половина штатов приходится на 2-7 млн, существуют некоторые выбросы с высоким населением. Верхняя и нижняя части коробки - это соответственно 75-й и 25-й процентиль или 3-й и 1-й квантиль. Усы коробки (линии от краев) сообщают размах для основной части данных. Обычно усы равняются 1.5 IQR - полуторам межквартальным размахам.

__Частотная таблица (frequency table)__ - сводка количеств числовых значений, которые разюиты на серию частотных интервалов (корзин, бинов). Частотная таблица переменной делит диапазон переменной на равноотстоящие сегменты и сообщает о том, сколько значений попадает в каждый сегмент.

```python
# pandas.cut - создает серию, которая увязывает значения с сегментами
binnedPopulation = pd.cut(state['Population'], 10)
binnedPopulation.value_counts()
```

Важно учитывать пустые корзины. Также полезно поэксперементировать с разными размерами корзинами. Если они слишком крупные, то важные признаки распределения могут быть затушеваны. Если же они слишком малые, то результат будет слишком гранулярным, и возможность наблюдать общую картину будет потеряна.

В общем случае квартили и децили будут иметь одинаковое количество в каждой корзине (равноколичественные корзины), но размеры корзин будут различаться. Частотная таблица, в отличие от них, будет иметь разные количества в корзинах (равноразмерные корзины), и размеры корзин будут одинаковыми.

__Гистограмма (histogram)__ - график частотной таблицы, в котором корзины откладываются на оси _х_, а количество - на оси _у_. Несмотря на внешнее сходство, столбиковые графики не следует путать с гистограммами. Гистограмма - это способ визуализации частотной таблицы. Столбики располагаются впритык друг к другу. Ось х представляет значения одной-единственной переменной на числовой шкале.

```python
# DataFrame.plot.hist имеет аргумент bins для определения числа корзин
# объект ax можно точнее отрегулировать визуализацию с помощью пакета matplotlib
ax = (state['Population'] / 1_000_000).plot.hist(figsize=(4, 4))
ax.set_xlabel('Численность населения')
```

__График плотности (density plot)__ - сглаженная версия гистограммы. График плотности вычисляется посредством _ядерной оценки плотности_. 

```python
# Пакет pandas метод density для создания плотностного графика
# аргумент bw_method для управления плавностью кривой плотностью

ax = state['Murder.Rate'].plot.hist(density=True, xlim=[0,12], bins=range(1,12))
state['Murder.Rate'].plot.density(ax=ax) # функция построения графика plot часто принимает необязательный аргумент ax
# что приводит к добавлению нового графика в тот же базовый график
ax.set_xlabel('Уровень убийств на 100тыс. населения')
```

Отличие от гистограммы в шкале оси _у_: график плотности соответствует отображению гистограммы как доли, а не количеств. Общая площадь под кривой плотности равна 1.

Методы оценки плотности в пакетах pandas и  scikit-learn предлагают хорошие имплементации.

## Разведывание категориальных данных

В случае категориальных данных представление о них дают простые доли или процентные соотношения.

__Мода (mode)__ - наиболее часто встречающаяся категория или значение в наборе данных. Обычно мода не используется для числовых данных.

__Ожидаемое значение (expected value)__ - (математическое ожидание) когда категории могут быть ассоциированны с числовыми значениями, этот показатель дает среднее значение значение, основываясь на вероятности появления категории. $$E\xi = {\sum_{i=1}^{n}\xi_ip_i}$$

__Столбиковые диаграммы (bar charts)__ - общепринятый визуальный инструмент для отображения одной-единственной категориальной переменной, где частота или доля каждой категории изображается в виде столбиков. Категории располагаются на оси _х_, частоты или доли на оси _у_.

```python
dfw = pd.read_csv('data/dfw_airline.csv')

ax = dfw.transpose().plot.bar(figsize=(4, 4), legend=False)
ax.set_xlabel('Причина задержки')
ax.set_ylabel('Количество')
```

На столбиковой диаграмме ось х представляет разные категории факторной переменной. Столбики располагаются отдельно друг от друга

__Круговые диаграммы (pie charts)__ - частота или доля каждой категории, изображаемая в виде сектора круга.

Часто столбиковые и круговые диаграммы относят к дезориентирующим графикам.

## Корреляция

__Коэффициент корреляции (correlation coefficient)__ - метрика, которая измеряет степень, с которой числовые переменные связаны друг с другом в диапазоне [-1, 1]. Переменные X и Y коррелируют положительно, если высокие значения Х сопровождаются высокими значениями Y и низкие значения Х сопровождаются низкими значениями Y. Если высокие значения Х сопровождаются низкими значениями Y, и наоборот, то переменные коррелируют отрицательно. Коэффициент корреляции чувствителен е выбросам в данных.

_Параметрический коэффициент корреляции Пирсона_, если перемнные имеют линейную связь: $r_{xy} = \frac{\sum_{i=1}^n(x_i - \bar{x})(y_i - \bar{y})}{(n-1)s_x s_y}$. Если перемнные взаимосвязаны нелинейно, то коэффициент корреляции Писона будет искажен. Также стоит использовать _коэффициент ранговой корреляции Спирмена и коэффициент ранговой корреляци Тенгеля $\tau$_ - эти коэффицменты робастны к выбросам и могут справояться с некоторыми типами нелинейности.

__Корреляционная матрица (correlation matrix)__ - таблица, в которой строки и столбцы - это переменные, а значенния ячеек - корреляции между этими переменными.

```python
sp500_sym = pd.read_csv('data/sp500_sectors.csv')
sp500_px = pd.read_csv('data/sp500_data.csv.gz', index_col=0)

telecomSymbols = sp500_sym[sp500_sym['sector'] == 'telecommunications_services']['symbol']

# Filter data for dates July 2012 through June 2015
telecom = sp500_px.loc[sp500_px.index >= '2012-07-01', telecomSymbols]
telecom.corr()
print(telecom)
```

```python
# тепловая карта с помощью функции heatmap пакета seaborn
etfs = sp500_px.loc[sp500_px.index > '2012-07-01', sp500_sym[sp500_sym['sector'] == 'etf']['symbol']]
sns.heatmap(etfs.corr(), vmin=-1, vmax=1, cmap=sns.diverging_palette(20, 220, as_cmap=True))
```

Методы в модуле sklearn.covariance пакета scikit-learn имплементируют целый ряд подходов - https://scikit-learn.org/stable/modules/classes.html#module-sklearn.covariance

__Диаграмма рассеяния (scatterplot)__ - график, в котором по оси _х_ откладываются значения одной переменной, а по оси _у_ - значения другой. Стандартным средством визуализации является _диаграмма рассеяния_.

```python
ax = telecom.plot.scatter(x='T', y='VZ', figsize=(4, 4), marker='$\u25EF$')
ax.set_xlabel('ATT (T)')
ax.set_ylabel('Verizon (VZ)')
ax.axhline(0, color='grey', lw=1)
ax.axvline(0, color='grey', lw=1)
```

## Разведывание двух или более переменных

Среднее значение и дисперсия рассматривают по одной переменной - __одномерный анализ__. Корреляционный анализ сравнивает две переменные - __двумерный анализ__. Но также можно прибегать к __многомерному анализу__.

__Таблица сопряженности (contigency table)__ - инструмент обобщения двух или более категориальных переменных, т.е.  сводка количеств двух или более категориальных переменных.

__Сетка из шестиугольных корзин (hexagonal binning)__ - график, показывающий плотность двух числовых переменных в форме топографической карты.

__Скрипичные графики (violin plots)__ - график, аналогичный коробчатой диаграмме, но показывающий оценку плотности. Скрипичный график представляет собой дополнение к коробчатой диаграмме и графически изображает оценки плотности. Преимущество скрипичного графика состоит в том, что он может показывать нюансы в распределении, которые не заметны на коробчатой диагрымме.

Тепловая карта, графики с сеткой из шестиугольных корзин и контурные графики дают визуальное представление о двумерной плотности, являются естественными аналогами гистограмм и графиков плотности.

```python
kc_tax = pd.read_csv('data/kc_tax.csv.gz')

kc_tax0 = kc_tax.loc[(kc_tax.TaxAssessedValue < 750000)&
                    (kc_tax.SqFtTotLiving > 100)&
                    (kc_tax.SqFtTotLiving < 3500), :]
kc_tax0.shape
```

```python
lc_loans = pd.read_csv('data/lc_loans.csv')
```

```python
# метод pivot_table создает сводную таблицу в Python
# аргумент aggfunc позволяет получить количества
crosstab = lc_loans.pivot_table(index='grade', columns='status', 
                                aggfunc=lambda x: len(x), margins=True) 
# аргумент margins добавит строчные и столбцовые суммы
print(crosstab)
```

```python
df = crosstab.copy().loc['A':'G',:] # копия сводной таблицы, игнорируя столбцовые суммы
df.loc[:,'Charged Off':'Late'] = df.loc[:,'Charged Off':'Late'].div(df['All'], axis=0) # делим строки на строчную сумму
df['All'] = df['All'] / sum(df['All']) # делим столбец 'All' на его сумму
perc_crosstab = df
print(perc_crosstab)
```

```python
# метод boxplot пакета pandas - коробчатые диаграмма
airline_stats = pd.read_csv('data/airline_stats.csv')
airline_stats.head()
ax = airline_stats.boxplot(by='airline', column='pct_carrier_delay',
                           figsize=(5, 5))
ax.set_xlabel('')
ax.set_ylabel('Daily % of Delayed Flights')
plt.suptitle('')

plt.tight_layout()
plt.show()
```

```python
# скрипичные графики строят с помощью violinplot пакета seaborn
fig, ax = plt.subplots(figsize=(5, 5))
sns.violinplot(data=airline_stats, x='airline', y='pct_carrier_delay',
               ax=ax, inner='quartile', color='white')
ax.set_xlabel('')
ax.set_ylabel('Доля ежедневных задержанных рейсов, %')
```

## Визуализация многочисленных переменных

Все виды графиков легко расширяемы на большее число переменных через понятие __кондиционности__. _Фасет (грань)_ - __кондиционные переменные (conditioning variables)__ - это фоновые переменные, характерезующие объект исследования, которые используются для построения правдоподобных значений путем привнесения недостающих данных.

Можно создавать фасетные графики с помощью matplotlib и seaborn. 

```python
zip_codes = [98188, 98105, 98108, 98126]
kc_tax_zip = kc_tax0.loc[kc_tax0.ZipCode.isin(zip_codes),:]
kc_tax_zip

def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=25, cmap=cmap, **kwargs)

g = sns.FacetGrid(kc_tax_zip, col='ZipCode', col_wrap=2) # аргументы col и row для указания кондиционных переменных
# для одной кондиционной переменной использовать аргумент col вместо col_wrap
# чтобы обернуть фасетные графики в несколько строк
g.map(hexbin, 'SqFtTotLiving', 'TaxAssessedValue', 
      extent=[0, 3500, 0, 700000]) # метод map вызывает функцию  hexbin с подмножествами исходного набора данных для
# разных почтовых индексов
# аргумент extent задает пределы осей Ox, Oy
g.set_axis_labels('Finished Square Feet', 'Tax Assessed Value')
g.set_titles('Zip code {col_name:.0f}')

plt.tight_layout()
plt.show()
```
